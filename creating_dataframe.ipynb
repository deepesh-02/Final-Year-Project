#**Mapping of filename to classid**

```python
import numpy as np
import pandas as pd
```

#**Importing the output labels**

```python
DATASET_PATH = 'drive/MyDrive/final_year_poject/dataset'
LABEL_PATH = 'drive/MyDrive/final_year_project/labels/output_label.txt'
```

```python
columns = ['class_id']
output_labels = pd.read_csv(LABEL_PATH, header=None, names=columns)
```

```python
output_labels.head()
```

#**Creating a dataframe for mapping filename to classid**

```python
mappings = pd.DataFrame(columns=['filename', 'class_id'])
```

```python
# Creating a filenames list

filenames = []
for i in range(len(output_labels['class_id'])):
  filenames.append(str(i + 1).zfill(4) + '.wav')
```

```python
filenames[1611]
```

```python
mappings['filename'] = filenames
mappings['class_id'] = output_labels['class_id']
```

```python
mappings
```

# **Exporting to csv**

```python
EXPORT_PATH = 'drive/MyDrive/final_year_project_e20-03/metadata/mappings.csv'
mappings.to_csv(EXPORT_PATH, header=None, index=None)
```
